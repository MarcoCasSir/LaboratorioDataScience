###  HOJA DE RUTA PARA EL PROYECTO
   - 1 Inicio y Configuracion
       * Introducccion: Breve descripción del proyecto (Predicción y clasificación de alquileres en Madrid).
       * Importacion de Librerias: Carga de Numpy, Pandas, Matplotlib, Seaborn, Scikit-learn y TensorFlow.
       * Carga del Dataset: Lectura del archivo CSV «Madrid Province Rent Data».

   - 2 Entendimientos de los Datos ( CRISP-DM)
       * Inspeccion Inicial: Uso de .info(), .describe() y .head() para revisar columnas como price, floor_area, bedrooms, etc..
       * Limpieza de Datos ( Data Wrangling):
             - Gestion de valores nuluos y duplicados
             - Conversion de tipos ( ej. pasar texto a número en columnas que lo requieran).
             - tratamiento de inconsistencias en year_built, floor o price.
    - 3 Analisis Exploratorio de datos (EDA)
      * Analisis Univariante: Histogramas de variables clave como el precio y el área.
      * Analisis Bivariante y Multivariante
              - Diagrama de dispersion ( ej: floor_ara vs price)
              - Uso de replot con parametros hue o size para ver realciones complejas
              - Matriz de correlacion para identificar dependencias.
      * Estudio de Atipicos ( Outliers): Identificacion de valores que puedan sesgar el modelo.
      * Estandarizacion para Comparacion: Comparar curvas de densidad en un dataframe copia.  
    - 4 Preparacion de los Datos con Pipelines
      * Definicion de Features (X) y target (y): Separacion para los dos casos ( Regresion para price y Clasidicacion para balcony).
      * Construccion del Pipeline de Preproceamiento:
              - Numericas: Imputacion de nulos, trasforamcion logaritmica( si hay asimetria) y escalado/normalizacion.
              - Categoricas: Codificacion ( ONe-Hot Encoding o Ordinal Encoding) para district, orientation, etc.
      * Division del Datse: Creacion de conjuntos de entrenamiento ( train)  y validacion ( test)  
    - 5 Modelado y Evaluacion
      * TAREA A : REGRESION ( Variable price)
              - Modelos de Scikit-learn (ej. LinearRegression, Random Forest). 
              - Red Neuronal con TensorFlow Keras.
              - Evaluación con métricas como el MSE.
      * TAREA A : REGRESION ( Variable price)
              - Modelos de Scikit-learn (ej. LogisticRegression, DecisionTree). 
              - Red Neuronal con TensorFlow Keras.
              - Evaluación con métricas como Accuracy.
    - 6 Conclusion y Documentacion
              - Justificación de las decisiones metodológicas tomadas.
              - Interpretación de los resultados obtenidos y su impacto potencial.
